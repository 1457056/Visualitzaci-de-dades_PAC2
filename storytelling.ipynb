```python
import pandas as pd

# Cargar el archivo CSV
data = pd.read_csv('/kaggle/input/accidents/US_Accidents_March23.csv')

# Obtener el número de filas
num_filas = data.shape[0]

print("El número de filas en el conjunto de datos es:", num_filas)
```

```python
import pandas as pd

# Obtener 20,000 filas aleatorias
sampled_data = data.sample(n=20000, random_state=42)

# Guardar el nuevo conjunto de datos reducido en un archivo CSV
sampled_data.to_csv('us-accidents-sampled.csv', index=False)

print("El nuevo conjunto de datos reducido ha sido guardado correctamente.")
```

```python
# Convertir la columna 'Start_Time' en formato de fecha
data['Start_Time'] = pd.to_datetime(data['Start_Time'])

# Extraer el año de la columna 'Start_Time'
data['Year'] = data['Start_Time'].dt.year

# Contar los accidentes por estado y por año
accidents_por_estado_y_año = data.groupby(['State', 'Year']).size().reset_index(name='Accident_Count')

# Pivotear los datos para tener los años como columnas
accidents_pivot = accidents_por_estado_y_año.pivot(index='State', columns='Year', values='Accident_Count')

# Guardar los datos en un archivo CSV
accidents_pivot.to_csv('accidents_por_estado_y_año.csv')
```

```python
import pandas as pd

# Convertir la columna 'Start_Time' en formato de fecha
data['Start_Time'] = pd.to_datetime(data['Start_Time'])

# Extraer la hora de la columna 'Start_Time'
data['Hour'] = data['Start_Time'].dt.hour

# Crear una columna para indicar si es de día o de noche
data['Day_Night'] = data['Hour'].apply(lambda hour: 'Day' if 6 <= hour < 18 else 'Night')

# Contar el recuento de accidentes por hora y por día/noche
recuento_accidentes_por_hora = data.groupby(['Hour', 'Day_Night'])['ID'].count().reset_index()
recuento_accidentes_por_hora.columns = ['Hour', 'Day_Night', 'Total_Accidents']

# Guardar los datos en un archivo CSV
recuento_accidentes_por_hora.to_csv('recuento_accidentes_por_hora.csv', index=False)

print("El recuento de accidentes por horas, separando día y noche, se ha guardado correctamente en un archivo CSV.")
```

```python
import pandas as pd
import calendar


# Convertir la columna 'Start_Time' en formato de fecha
data['Start_Time'] = pd.to_datetime(data['Start_Time'])

# Extraer el día y el mes de la columna 'Start_Time' como nombres
data['Day'] = data['Start_Time'].dt.strftime('%A')
data['Month'] = data['Start_Time'].dt.strftime('%B')

# Definir el orden personalizado para los días de la semana
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Definir el orden personalizado para los meses
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Mapear los días y los meses al orden personalizado
day_mapping = {day: i for i, day in enumerate(day_order)}
month_mapping = {month: i for i, month in enumerate(month_order)}

data['Day'] = data['Day'].map(day_mapping)
data['Month'] = data['Month'].map(month_mapping)

# Ordenar el dataset por días de la semana y meses
data_sorted = data.sort_values(by=['Day', 'Month'])

# Restaurar los nombres de los días y los meses
data_sorted['Day'] = data_sorted['Day'].map({v: k for k, v in day_mapping.items()})
data_sorted['Month'] = data_sorted['Month'].map({v: k for k, v in month_mapping.items()})

# Contar el recuento de accidentes por día y mes
recuento_accidentes_por_dia_mes = data_sorted.groupby(['Day', 'Month']).size().reset_index(name='Total_Accidents')

# Guardar los datos en un archivo CSV
recuento_accidentes_por_dia_mes.to_csv('recuento_accidentes_por_dia_mes.csv', index=False)

```

```python
# Agrupar los datos por precipitación y contar el número de accidentes en cada categoría
recuento_accidentes_por_precipitacion = data.groupby('Precipitation(in)').size().reset_index(name='Total_Accidents')

# Ordenar los resultados por precipitación de manera ascendente
recuento_accidentes_por_precipitacion = recuento_accidentes_por_precipitacion.sort_values(by='Precipitation(in)', ascending=True)

# Guardar los datos en un archivo CSV
recuento_accidentes_por_precipitacion.to_csv('recuento_accidentes_por_precipitacion.csv', index=False)
```

```python
# Definir los límites de los intervalos y las etiquetas de las categorías
bins = [0, 0.01, 0.1, float('inf')]
labels = ['Dry', 'Wet', 'Rain']

# Segmentar la columna "Precipitation(in)" en categorías
data['Precipitation_Category'] = pd.cut(data['Precipitation(in)'], bins=bins, labels=labels, right=False)

# Calcular el recuento de accidentes por categoría de precipitación
recuento_accidentes_por_precipitacion = data['Precipitation_Category'].value_counts().reset_index()
recuento_accidentes_por_precipitacion.columns = ['Categoria_Precipitacion', 'Total_Accidents']

# Guardar los datos en un archivo CSV
recuento_accidentes_por_precipitacion.to_csv('recuento_accidentes_por_precipitacion.csv', index=False)
```
